<h2> a) Construcción Dataframe </h2>


```python
import pandas as pd
import numpy as np
import copy

url = 'http://statweb.stanford.edu/~tibs/ElemStatLearn/datasets/prostate.data'

def getTrainSetData(url):

   df = pd.read_csv(url, sep='\t', header=0) #Read the CSV file
   df = df.drop('Unnamed: 0', axis=1) # removes a set of labels from an axis, first column
   istrain_str = df['train']   #get the train column
   istrain = np.asarray([True if s == 'T' else False for s in istrain_str])
   istest = np.logical_not(istrain)
   df = df.drop('train', axis=1) #Drop  It removes a set of labels from an axis, last column
   return df, istrain , istest

df, istrain , istest = getTrainSetData(url)
```

La linea <code>"df = df.drop('Unnamed: 0') ,axis=1)"</code> remueve las etiquetas correspondientes al índice de cada fila de datos del Dataframe, es decir elimina la primera columna del set de datos.

Por otro lado <code>"df = df.drop('train', axis=1) "</code> elimina la última columna que permite diferenciar entre los datos que corresponden al set de entrenamiento y el set de pruebas.

<h2> c) Normalización de Datos </h2>

```python
from sklearn.preprocessing import StandardScaler

def ScaleData(df):
   scaler = StandardScaler()
   df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
   df_scaled['lpsa'] = df['lpsa']
   return df_scaled

df_scaled = ScaleData(df)
```

La importancia de realizar una normalización de datos es que dado el hecho de que estamos trabajando con múltiples atributos (9 en este caso) , necesitamos que estos se encuentren en una escala similar o un rango de valores similares, de lo contrario, el tiempo necesario para encontrar los coeficientes óptimos al aplicar regresión lineal con <em> Gradient Descent </em> por ejemplo, será mucho mayor comparado los atributos en un intervalo similar.

Las técnicas comunes para normalizar los datos son <em> Feature Scaling </em> y <em> Mean Normalization </em> 


<h3> Regresión Lineal </h3>


```python
import sklearn.linear_model as lm
def getXdata(df_scaled):
   X = df_scaled.ix[:,:-1]
   N = X.shape[0]
   X.insert(0, 'intercept', np.ones(N))
   Xtrain , Xtest = X[istrain] , X[np.logical_not(istrain)]
   
   return Xtrain , Xtest

def getydata(df_scaled):
   y = df_scaled['lpsa']
   ytrain , ytest = y[istrain] , y[np.logical_not(istrain)]
   return ytrain , ytest 


def getLinearRegression(df_scaled):
   X = df_scaled.ix[:,:-1]
   y = df_scaled['lpsa']

   N = X.shape[0]
   X.insert(0, 'intercept', np.ones(N))
   
   Xtrain , xtest=  getXdata(df_scaled)
   ytrain , ytest = getydata(df_scaled)
   
   linreg = lm.LinearRegression(fit_intercept = False)
   linreg.fit(Xtrain, ytrain)

   return linreg

def getStandardRegressionError(linreg, Xtrain , ytrain ):
   std_reg = np.sqrt(((linreg.predict(Xtrain) - ytrain) ** 2).sum()/58)
   return std_reg

def getZscore(linreg,std_reg,K):
   inverse_covariance = np.linalg.inv(Xtrain.transpose().dot(Xtrain))
   z_score = np.ones(K)
   for i in range(0,K):
      z_score[i] = linreg.coef_[i]/(std_reg *np.sqrt((inverse_covariance[i,i])))
   return z_score 

linreg = getLinearRegression(df_scaled)
```

¿Como podemos verificar si una variable está relacionada con la respuesta? podemos implementar la <strong> hipótesis </strong> de eliminar cierto atributo, es decir es una hipótesis de igualdad $B_{j}= 0$, la cual puede probarse calculando los Z-score estandarizados: 


$z_{j}= \frac{B_{j}}{\lambda \sqrt{v_{j}}}$

Bajo la hipotesis nula de igualdad $H_{0} :B _{j}= 0$, el coeficiente $z_{j}$ está distribuido como una distribución $t$ con $(N-p-1)$ grados de libertad(<em> two-tailed t distribution </em>). Bajo ese análisis, todo  valor absoluto de z-score considerablemente mayor a $2.0002$ conllevará a un rechazo de la hipótesis nula, usando un nivel de significación de $5%$. Observando la tabla, el atributo <strong> lcavol</strong> muestra el z-score mas elevado, acompañado de <strong> lweight </strong> y <strong> svi </strong>, con $2.7505$ y $2.4693$ respectivamente.


<h2> f) Error de Predicción del Modelo </h2>

```python
from sklearn import cross_validation

def getKfold(linreg,Xtest,ytest,Xtrain,ytrain, K):
   yhat_test = linreg.predict(Xtest)
   mse_test = np.mean(np.power(yhat_test - ytest, 2))

   Xm = Xtrain.as_matrix()
   ym = ytrain.as_matrix()

   k_fold = cross_validation.KFold(len(Xm), K)
   mse_cv = 0
   for k , (train, val) in enumerate(k_fold):
      linreg = lm.LinearRegression(fit_intercept = False)
      linreg.fit(Xm[train] , ym[train])
      yhat_val = linreg.predict(Xm[val])
      mse_fold = np.mean(np.power(yhat_val - ym[val], 2))
      mse_cv += mse_fold
   mse_cv = mse_cv/K

   return mse_cv

Xtrain , Xtest=  getXdata(df_scaled)
ytrain , ytest = getydata(df_scaled)

print "-- K-Fold --"
print "K = 5 : " + str(getKfold(linreg,Xtest,ytest,Xtrain,ytrain, 5))
print "K = 10 : " + str(getKfold(linreg,Xtest,ytest,Xtrain,ytrain, 10))
print "Residual sum of squares : " + str( np.mean((linreg.predict(Xtest) - ytest) ** 2))
```

Se observa que usando <em>cross-validation</em> en los datos, el error es considerablemente mayor en comparación a imeplementar el test sin <em>K-fold</em>. Esto revela que los datos usados dependen del modelo, y no logra realizar predicciónes que sean confiables con nuevos datos externos al set de entrenamiento  ( <strong> Overfitting </strong> ). Este fenómeno suele ocurrir cuando el modelo que se busca generar es complejo, es decir, hay muchos parámetros y una de las soluciones para reducir el overffitng puede ser o disminuir el número de atributos en el modelo o conseguir una mayor cantidad de datos de entrenamiento.

<h2> QQ Plot </h2>


```python
%matplotlib inline
import scipy.stats as stats
import matplotlib.pyplot as plt

def getQQPlot(Xtrain,ytrain):
   predict = linreg.predict(Xtrain)
   ytrain = np.asarray(ytrain)
   prediction_error = predict - ytrain 

   
   stats.probplot(prediction_error, dist='norm', plot=plt)
   plt.ylabel('Training Data Error')
   plt.title('Quantile-Quantile Plot')
   plt.show()

getQQPlot(Xtrain,ytrain)
```

<h1>Selección de Atributos </h1>

<h2>a) Forward Step-Wise Selection </h2>

A veces , la implementación de mínimos cuadrado puede conllevar a una alta varianza, y esto puede mejorarse penalizando, o disminuyendo algunos coeficientes del modelo a cero, sacrificando algo de parcialidad con el objetivo de mejorar el performance general del modelo. Con <strong> Forward Step-Wise Selection </strong> buscamos ir agregando iterativamente al modelo cada atributo, basandonos en un score de cada candidato. Se analizarán distintos criterios para analizar el mejor candidato en cada fase incremental del modelo.

<h3>Criterios:</h3>
1) Error cuadrático medio. <br><br>
$\frac{1}{n} \sum ^{n}_{i=1}|e^{i}|$ <br>
<br>
2) Error absoluto medio<br><br>
$\frac{1}{n} \sum ^{n}_{i=1} (y_{predict} - y_{i})^2$ <br>
<br>

3) Z-score<br><br>

Por otro lado, en términos de interpretación,a veces buscamos determinar un subset pequeño que logre el efecto mas fuerte en optimización de predicción, en otras palabras, buscamos entender la "imagen general" , sacrificando los detalles mas pequeños (atributos que no interfieren mucho con el resultado final)

```python
names_regressors = ["Lcavol","Lweight","Age","Lbph","Svi","Lcp","Gleason","Pgg45"]

#
#Forward Step-Wise Selection with Mean squared Error
#
def fss_mse( x , y , xtest, ytest, names_x, k = 10000):
   p = x.shape[1] - 1
   k = min(p,k)
   names_x = np.array(names_x)
   remaining = range(1,p+1)
   training_error = []
   test_error = []
   selected = [0]
   current_score = 0.0
   best_new_score = 0.0

   while remaining and len(selected) <= k :
      score_candidates = []
      for candidate in remaining:
         model = lm.LinearRegression(fit_intercept = False)
         indexes = selected + [candidate]

         x_train = x[:,indexes]
         x_test = xtest[:,indexes]

         predictions_train = model.fit(x_train,y).predict(x_train)
         residuals_train = predictions_train - y

         predictions_test = model.fit(x_train,y).predict(x_test)
         residuals_test= predictions_test - ytest
         mse_train_candidate = np.mean(np.power(residuals_train,2))
         mse_test_candidate = np.mean(np.power(residuals_test,2))
         score_candidates.append((mse_train_candidate, mse_test_candidate ,candidate))

      score_candidates.sort()
      score_candidates[:] = score_candidates[::-1]
      best_new_score , test_score, best_candidate = score_candidates.pop()
      training_error.append(best_new_score)
      test_error.append(test_score)
      remaining.remove(best_candidate)
      selected.append(best_candidate)
      print  "selected = %s ..." %names_x[best_candidate-1]
      print  "totalvars = %d, mae = %f " %(len(indexes), best_new_score)
   return selected , training_error, test_error

#
#Forward Step-Wise Selection with Mean Absolute Error
#
def fss_mae( x , y , xtest, ytest, names_x, k = 10000):
   p = x.shape[1] - 1
   k = min(p,k)
   names_x = np.array(names_x)
   remaining = range(1,p+1)
   training_error = []
   test_error = []
   selected = [0]
   current_score = 0.0
   best_new_score = 0.0

   while remaining and len(selected) <= k :
      score_candidates = []
      for candidate in remaining:
         model = lm.LinearRegression(fit_intercept = False) 
         indexes = selected + [candidate]
         x_train = x[:,indexes]
         x_test = xtest[:,indexes]

         predictions_train = model.fit(x_train,y).predict(x_train)
         residuals_train = np.fabs(predictions_train - y)

         predictions_test = model.fit(x_train,y).predict(x_test)
         residuals_test = np.fabs(predictions_test - ytest)

         mae_training_candidate = np.mean(residuals_train)
         mae_test_candidate = np.mean(residuals_test)
         score_candidates.append((mae_training_candidate,mae_test_candidate , candidate))

      score_candidates.sort()
      score_candidates[:] = score_candidates[::-1]
      best_new_score , test_score ,best_candidate = score_candidates.pop()
      training_error.append(best_new_score)
      test_error.append(test_score)
      remaining.remove(best_candidate)
      selected.append(best_candidate)
      print  "selected = %s ..." %names_x[best_candidate-1]
      print  "totalvars = %d, mse = %f " %(len(indexes), best_new_score)
   return selected , training_error, test_error
#
#Forward Step-Wise Selection with Z-score
#
def fss_zscore( x , y , xtest, ytest, names_x, k = 10000):
   p = x.shape[1] - 1
   k = min(p,k)
   names_x = np.array(names_x)
   remaining = range(1,p+1)
   selected = [0]
   training_error = []
   test_error = []
   current_score = 0.0
   best_new_score = 0.0

   while remaining and len(selected) <= k :
      score_candidates = []
      for candidate in remaining:
         model = lm.LinearRegression(fit_intercept = False)
         indexes = selected + [candidate]
         
         x_train = x[:,indexes]
         x_test = xtest[:,indexes]
         
         linreg = model.fit(x_train,y)
         std_reg = getStandardRegressionError(linreg, x_train , y )
         
         zscore_candidate = abs(getZscore(linreg,std_reg,len(selected)+1)[-1])

         predictions_train = model.fit(x_train,y).predict(x_train)
         residuals_train = predictions_train - y

         predictions_test = model.fit(x_train,y).predict(x_test)
         residuals_test = predictions_test - ytest

         zscore_train_residual = np.mean(np.power(residuals_train,2))
         zscore_test_residual = np.mean(np.power(residuals_test,2))
   
         score_candidates.append((zscore_candidate,zscore_train_residual,zscore_test_residual ,candidate))

      score_candidates.sort()
      best_new_score , train_residual, test_residual, best_candidate = score_candidates.pop()
      training_error.append(train_residual)
      test_error.append(test_residual)
      remaining.remove(best_candidate)

      selected.append(best_candidate)
      print  "selected = %s ..." %names_x[best_candidate-1]
      print  "totalvars = %d, zscore = %f " %(len(indexes), best_new_score)
   return selected , training_error, test_error

def plotFSS(train_error,test_error, method):
    p = range(1,9)
    plt.plot(p,train_error, color="green", linewidth=3, label = "Training Set")
    plt.plot(p,test_error, color="red",linewidth=3, label = "Testing set")
    plt.legend()
    axes = plt.gca()
    if method == "mse":
        plt.xlabel('Numero de variables')
        plt.ylabel('Error MSE')
        plt.title('FFS MSE ')
        axes.set_ylim([0.3,0.8])
    
    if method == "mae":
        plt.xlabel('Numero de variables')
        plt.ylabel('FFS MAE')
        plt.title('FFS MAE')
        axes.set_ylim([0.4,0.7])
        
    if method == "zscore":
        plt.xlabel('Numero de variables')
        plt.ylabel('FFS Z-score')
        plt.title('FFS Zscore')
        axes.set_ylim([0.3,0.8])
    plt.show()

selected_mse , train_error_mse , test_error_mse  = fss_mse( Xtrain.as_matrix() ,ytrain.as_matrix() , Xtest.as_matrix() ,ytest.as_matrix() ,names_regressors )

selected_mae , train_error_mae, test_error_mae = fss_mae( Xtrain.as_matrix() ,ytrain.as_matrix() , Xtest.as_matrix() ,ytest.as_matrix() ,names_regressors )

selected_z , train_error_z, test_error_z = fss_zscore( Xtrain.as_matrix() ,ytrain.as_matrix() , Xtest.as_matrix() ,ytest.as_matrix() ,names_regressors )

plotFSS(train_error_mse , test_error_mse, "mse")
plotFSS(train_error_mae , test_error_mae, "mae")
plotFSS(train_error_z , test_error_z, "zscore")
```

<h2> b.) Backward Step-wise Selection </h2>

```python
#------------------------------------------------
#BSS Mean Squared Error
#----------------------------------------
def bss_mse( x , y , xtest, ytest, names_x, k = 10000):
   p = x.shape[1] 
   k = min(p,k) 
   names_x = np.array(names_x)
   remaining = range(0,p)
   selected = range(0,p)
   training_error = []
   test_error = []
   current_score = 0.0
   best_new_score = 0.0

   while len(remaining) > 1 and len(selected) <= k :
      score_candidates = []
      for candidate in remaining:
         model = lm.LinearRegression(fit_intercept = False)
         indexes = copy.copy(selected) 
         indexes.remove(candidate)
         x_train = x[:,indexes]
         x_test = xtest[:,indexes]

         predictions_train = model.fit(x_train,y).predict(x_train)
         residuals_train = predictions_train - y

         predictions_test = model.fit(x_train,y).predict(x_test)
         residuals_test = predictions_test - ytest

         mse_train_candidate = np.mean(np.power(residuals_train,2))
         mse_test_candidate = np.mean(np.power(residuals_test,2))

         score_candidates.append((mse_train_candidate, mse_test_candidate, candidate))
         
      score_candidates.sort()
      score_candidates[:] = score_candidates[::-1]
      best_new_score , test_score, best_candidate = score_candidates.pop()
      training_error.append(best_new_score)
      test_error.append(test_score)
      remaining.remove(best_candidate)
      selected.remove(best_candidate)
      print  "selected = %s ..." %names_x[best_candidate-1]
      print  "totalvars = %d, mse = %f " %(len(indexes), best_new_score)
      
   return selected , training_error, test_error

#------------------------------------------------
#BSS Mean Absolute Error
#------------------------------------------------
def bss_mae( x , y , xtest, ytest, names_x, k = 10000):
   p = x.shape[1] 
   k = min(p,k) 
   names_x = np.array(names_x)
   remaining = range(0,p)
   selected = range(0,p)
   training_error = []
   test_error = []
   current_score = 0.0
   best_new_score = 0.0

   while len(remaining) > 1 and len(selected) <= k :
      score_candidates = []
      for candidate in remaining:
         model = lm.LinearRegression(fit_intercept = False)
         indexes = copy.copy(selected) 
         indexes.remove(candidate)
         x_train = x[:,indexes]
         x_test = xtest[:,indexes]

         predictions_train = model.fit(x_train,y).predict(x_train)
         residuals_train = np.fabs(predictions_train - y)

         predictions_test = model.fit(x_train,y).predict(x_test)
         residuals_test = np.fabs(predictions_test - ytest)


         mae_train_candidate = np.mean(residuals_train)
         mae_test_candidate = np.mean(residuals_test)

         score_candidates.append((mae_train_candidate, mae_test_candidate, candidate))
         
      score_candidates.sort()
      score_candidates[:] = score_candidates[::-1]
      best_new_score , test_score,best_candidate = score_candidates.pop()
      training_error.append(best_new_score)
      test_error.append(test_score)
      remaining.remove(best_candidate)
      selected.remove(best_candidate)
      print  "selected = %s ..." %names_x[best_candidate-1]
      print  "totalvars = %d, mse = %f " %(len(indexes), best_new_score)
      
   return selected , training_error, test_error

def plotBSS(train_error,test_error, method):
    p = range(1,9)
    plt.plot(p,train_error, color="blue", linewidth=3, label = "Training Set")
    plt.plot(p,test_error, color="red",linewidth=3, label = "Testing set")
    plt.legend()
    axes = plt.gca()
    plt.gca().invert_xaxis()
    if method == "mse":
        plt.xlabel('Numero de variables')
        plt.ylabel('Error MSE')
        plt.title('BBS MSE ')
        axes.set_ylim([0.2,1.4])
    
    if method == "mae":
        plt.xlabel('Numero de variables')
        plt.ylabel('Error MAE')
        plt.title('BBS MAE')
        axes.set_ylim([0.2,1.4])
        
    plt.show()

selected_mse , train_error_mse , test_error_mse  = bss_mse( Xtrain.as_matrix() ,ytrain.as_matrix() , Xtest.as_matrix() ,ytest.as_matrix() ,names_regressors )

selected_mae , train_error_mae, test_error_mae = bss_mae( Xtrain.as_matrix() ,ytrain.as_matrix() , Xtest.as_matrix() ,ytest.as_matrix() ,names_regressors )

plotBSS(train_error_mse , test_error_mse, "mse")
plotBSS(train_error_mae , test_error_mae, "mae")
```

<h1> 3) Regularización </h1>

Otra forma de reducir el <strong>overffit</strong> en modelos complejos , es penalizar los atributos, reduciendo o eliminando usando coeficiente de truncamiento $\lambda$.Los métodos a analizar a continuación son <strong>Ridge Regression</strong> y <strong>Lasso</strong>  

<h2> Ridge Regression </h2>


Ridge Regression reduce los coeficientes de regresión con una variable $\lambda$ ,imponiendo un penalty al valor de cada coeficiente, mientras mas grande sea $\lambda$ , mayor será la reducción, por lo que hay buscar un valor de $\lambda$ eficiente, dado que si $\lambda$ es muy grande, el modelo puede sufrir <strong>underfitting</strong> o un bias elevado, mientras que un $\lambda$ demasiado bajo, puede no afectar al modelo en absoluto y no observar cambios en el <strong>overffiting</strong>.
<br>
<br>
Para aplicar Ridge Regressión es necesario que los datos estén estandarizados y eliminar el intercepto en la penalización, dado que si se penaliza el intercepto, se generará un procedimiento dependiente del origen del eje $Y$

```python
plt.figure(figsize=(12,7))
from sklearn.linear_model import Ridge

def ridge_Regression( Xtrain , ytrain , names_regressors ):

   Xtrain = Xtrain.drop('intercept',axis = 1)
   alphas_ = np.logspace(4,-1,base=10)
   coefs = []
   model = Ridge(fit_intercept = True,solver='svd')
   for a in alphas_:
      model.set_params(alpha=a)
      model.fit(Xtrain, ytrain)
      coefs.append(model.coef_)
   ax = plt.gca()
   
   for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
      plt.plot(alphas_, y_arr, label=label)
        
   plt.legend()
   ax.set_xscale('log')
   ax.set_xlim(ax.get_xlim()[::-1]) # reverse axis
   plt.xlabel('alpha')
   plt.ylabel('weights')
   plt.title('Regularization Path RIDGE')
   plt.axis('tight')
   plt.legend(loc=2)
   plt.show()

ridge_Regression(Xtrain,ytrain,names_regressors )
```

Ridge Regression genera una reducción proporcional, lo cual no funciona bien antes modelos no lineales. Con un valor de $\lambda$ cercano a cero, existe una diferencia considerable entre cada atributo, proporcional a los zscore obtenidos anteriormente, siendo <em> Lcavol </em> nuevamente predominante. Conforme $\lambda$ va aumentando, se observa la reducción de los coeficientes, pero Lcavol se mantiene constantemente como el coeficiente de mayor peso. Para un $\lambda$ cercano a $10^4$ el modelo presenta underfit, es decir una parcialidad elevada, por lo que el modelo no podrá generar predicciones correctamente.

<h2> b.) Lasso </h2>

A diferencia de Ridge Regression, <strong> Lasso </strong> interviene en cada coeficiente por un factor de $\lambda$, truncando a cero cuando no se cumple la condición , esto genera soluciones no lineales y consiste en un problema de programación cuadrática.
<br>
<br>


```python
plt.figure(figsize=(12,7))
from sklearn.linear_model import Lasso
def theLasso(Xtrain,ytrain,names_regressors):

   Xtrain = Xtrain.drop('intercept',axis = 1)
   alphas_ = np.logspace(1,-5,base=10)
   coefs = []
   model = Lasso(fit_intercept=True)
   for a in alphas_:
      model.set_params(alpha=a)
      model.fit(Xtrain, ytrain)
      coefs.append(model.coef_)
   ax = plt.gca()
   #print np.squeeze(coefs).T
   for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
      plt.plot(alphas_, y_arr, label=label)
   plt.legend()
   ax.set_xscale('log')
   ax.set_xlim(ax.get_xlim()[::-1]) # reverse axis
   plt.xlabel('alpha')
   plt.ylabel('weights')
   plt.title('Regularization Path LASSO')
   plt.axis('tight')
   plt.legend(loc=2)
   plt.show()

theLasso(Xtrain,ytrain,names_regressors )
```

Lasso es mucho mas estricto e interpretativo que Ridge Regression, el punto de que algunos atributos son igualmente descartados con un $\lambda$ relativamente cercano a cero, y por ello mismo el gráfico fue analizado con un intervalo de valores lambda mas reducidos que con Ridge Regression.
<br>
<br>

Comparando los atributos afectados, las variables más relevantes son nuevamente <strong>Lcavol</strong> , <strong>Lweight</strong> y <strong>Svi</strong>  ,favoreciendolos dado su mayor variabilidad. En este caso fue necesario un valor de $\lambda$ considerablemente bajo( menor que $10^{-2}$) para que el atributo <strong>Gleason</strong> no fuera penalizado, confirmando su poca relevancia en el modelo final. Otros atributos de poca relevancia son <strong>Age</strong> y <strong>Lcp</strong>

```python
def MSE(y,yhat): return np.mean(np.power(y-yhat,2))

def crossValidationParameterRidge(Xtrain,ytrain):
    alphas_ = np.logspace(2,-2,base=10)

    k_fold = cross_validation.KFold(len(Xtrain),10)
    best_cv_mse = float("inf")
    model = Ridge(fit_intercept = True,solver='svd')
    for a in alphas_:
        model.set_params(alpha=a)
        mse_list_k10 = [MSE(model.fit(Xtrain[train], ytrain[train]).predict(Xtrain[val]), ytrain[val]) for train, val in k_fold]
        if np.mean(mse_list_k10) < best_cv_mse:
            best_cv_mse = np.mean(mse_list_k10)
            best_alpha = a
            print "BEST PARAMETER = %f, MSE(CV) = %f"%(best_alpha,best_cv_mse)
            

def crossValidationParameterLasso(Xtrain,ytrain):
    alphas_ = np.logspace(0.5,-2,base=10)

    k_fold = cross_validation.KFold(len(Xtrain),10)
    best_cv_mse = float("inf")
    model = Lasso(fit_intercept=True)
    for a in alphas_:
        model.set_params(alpha=a)
        mse_list_k10 = [MSE(model.fit(Xtrain[train], ytrain[train]).predict(Xtrain[val]), ytrain[val]) for train, val in k_fold]
        if np.mean(mse_list_k10) < best_cv_mse:
            best_cv_mse = np.mean(mse_list_k10)
            best_alpha = a
            print "BEST PARAMETER = %f, MSE(CV) = %f"%(best_alpha,best_cv_mse)
            
print"-----Cross Validation Parameter----- \n"
crossValidationParameterRidge(Xtrain.as_matrix(),ytrain.as_matrix())

print"-----Cross Validation Parameter----- \n"
crossValidationParameterLasso(Xtrain.as_matrix(),ytrain.as_matrix())
```

```python
import pandas as pd
import numpy as np
import sklearn.linear_model as lm
from scipy.sparse import csr_matrix
from scipy.io import mmread
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

Cargamos el dataset de entrenamiento, validación y testeo. Se utiliza el tipo de matrices sparse por su alto contenido de celdas nulas. Este tipo de matrices permiten utilizar algoritmos especiales para agilizar las operaciones que se hacen con ellas.

```python
# training set
X_train = csr_matrix(mmread('train.x.mm'))
y_train = np.loadtxt('train.y.dat')
# test set
X_test = csr_matrix(mmread('test.x.mm'))
y_test = np.loadtxt('test.y.dat')
# CV
X_cv = csr_matrix(mmread('dev.x.mm'))
y_cv = np.loadtxt('dev.y.dat')
```

Cargando los datos de entrenamiento se procede a generar el modelo de regresión lineal obteniendo R2 o coeficiente de determinación

```python
model = lm.LinearRegression(fit_intercept = False)
model.fit(X_train,y_train)
print "Regresion lineal\nCV: ",model.score(X_cv,y_cv),"\nTest: ",model.score(X_test, y_test)
```

Regresión lineal tiene un $R^2=0.612862766741$ para el set de validación.
En el paper "Movie Reviews and Revenues: An Experiment in Text Regression" de Mahesh Joshi et al., utilizan para resolver este problema una mezcla entre normas $\ell_1$ y $\ell_2$ (ridge y lasso) para la regularización de datos llamada Elastic Net. Presenta un trade-off entre las regularizaciones anteriores determinada por una valor $\alpha\in(0,1)$. Elastic net es útil cuando hay muchas características relacionadas unas con otras. Utilizando una lista de valores de alpha entre 0 y 1, con un trade-off de 0.9, los resultados son los siguientes.

```python
alpha = [0.1, 0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1.0]
cr2 = []
tr2 = []

for a in alpha:
	# modelo con elasticnet
	# se calcula el valor de R2 o coeficiente de determinacion
	enet = lm.ElasticNet(fit_intercept = False)
	enet.set_params(alpha= a, l1_ratio=0.9, max_iter = 1000)
	enet.fit(X_train, y_train)
	
	cvR2 = enet.score(X_cv,y_cv)
	testR2 = enet.score(X_test, y_test)
	cr2.append(cvR2)
	tr2.append(testR2)
	print "alpha: ",a,"\nCV: ",cvR2,"\nTest: ",testR2
```

El mejor resultado obtenido con ElasticNet es un $R^2=0.620626081934$ con los parámetros $alpha=0.25$ y $l1\_ratio=0.9$, siendo mayor al de regresión lineal. Dado ese valor de coeficiente de determinación, sacando la raiz cuadrada se obtiene un coeficiente de correlación de Pearson de 0.78779824951.
